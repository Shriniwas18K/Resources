###### whenever we take out single row then we get series, else multiple rows are dataframe , we can use series method of loc iloc fancy indexing to get rows columns anything from dataframe , similar to list slicing

```python
ipl.iloc[9:15:3]
```

##### major diffirence in iloc and loc is that : iloc doesnot include upper bound , whereas loc includes upper bound

```python
ipl.loc[9:15:3]
```

##### loc is used to access the custom given index , and iloc is used to access default indexing inside dataframe : the dataframe always contains its default indexing and we can use iloc and loc both to access it irrespective of we have given custom  indexing or not
##### iloc cannot be used on custom indexing made by set_index method, we need to use loc only for custom index

```python
temp=ipl.set_index('mid')
temp.loc[1359475:1359477]
```
