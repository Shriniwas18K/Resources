### the DataFrames that we get from groupby are all Multiindex DataFrames
### MultiIndex DataFrames are used to represent higher dimension data into 2D - DataFrame
### A Data which has large number of features is called higher dimensional data

### Basically we are forming a tree Data Structure whose leaf nodes are individual data points.This might not be understood without the video . so pls go and watch the MultiIndex DataFrame video

```python
import numpy as np
import pandas as pd
```

##### to begin with we will start with MultiIndex Series: 
##### note : multiindex object cannot contain duplicate values of index , as it is actually index , i.e. unique identifier for something so it should not contain duplicate values , so while even making it we have not given below any duplicate values in below cell

```python
a=['p','n']
b=[123,4,554,523]
l=['india','uk','aus']
```

##### we use pd.MultiIndex.from_product([a,b,l]) to make MultiIndex series : it is basically cross product as a x b x l

```python
mulind=pd.MultiIndex.from_product([a,b,l])
mulind
```

##### similarly we can have many other lists in that cross product , this method creates a MultiIndex object

##### now we can create a series which can have indices as that of this MultiIndex object , but the shape of MultiIndex object ,and series should be same, as in series no value can exist without its index

```python
mulind.shape
```

```python
arr=np.random.randint(10,100,24)
s=pd.Series(arr,index=mulind)
s
```

##### this when seen horizontally forms tree data structure

##### here [ ' p ' , ' n ' ] is level 0 row feature where 'p' and 'n' are nodes , similarly [ 123 , 4 , 554 , 523 ] is level 1 row feature which contains inside nodes , and [ ' india ' , ' uk ' , ' aus ' ] is level 2 row feature

#### now we will make a MultiIndex DataFrame , and put its index as that multiindex object

```python
arr=np.random.randint(10,100,96).reshape(24,4)
df=pd.DataFrame(arr,index=mulind,columns=['a','b','c','d'])
df
```

##### actually internally pandas treats row indices and column names internally as numpy arrays,so column names can also be MultiIndex objects , so now we will columns as multiindex object , but shape needs to match

```python
y=['a','b','c']
z=[10,'t','r']
mulindcol=pd.MultiIndex.from_product([z,y])
```

```python
mulind.shape
```

```python
mulindcol.shape
```

```python
df=pd.DataFrame(np.random.randint(10,200,216).reshape(24,9),index=mulind,columns=mulindcol)
df
```

##### from above dataframe [ 10 , ' t ' , ' r ' ] are level 0 features , [ ' a ' , ' b ' , ' c ' ] are level 1 features for columns 
##### here [ ' p ' , ' n ' ] are level 0 features, similarly [ 123 , 4 , 554 , 523 ] are level 1 features, and [ ' india ' , ' uk ' , ' aus ' ] are level 2 features of rows
##### currently [ ' a ' , ' b ' , ' c ' ] is innermost level column feature , [ ' india ' , ' uk ' , ' aus ' ] is innermost level row feature

#### this above dataframe represents 5D - dimensional data in a two-dimensional DataFrame , as value of any cell depends upon those 5 features for a data point in above DataFrame . for example consider exactly first cell with value 41  , it is actually as df[10]['a']['p'][123]['india']=41

##### now we can access the values of the DataFrame using rows and column names as well as normal way of iloc loc where we use 0-based indexing, as behind the scenes pandas still maintains internally 0-based indexing

```python
df[10]['a']
```

```python
df[10]['a']['n']
```

```python
df[10]['a']['n'][4]
```

```python
df[10]['a']['n'][4]['aus']
```

##### this way we access starting from level 0 to innermost column feature , and then level 0 to innermost row feature

```python
df[:10:2]
```

```python
df.iloc[:5]
```

#### we have two methods unstack and stack
#### unstack converts innermost row feature to innermost column feature
#### stack converts innermost column feature to innermost row feature
#### both work keeping the existing features
#### when no features are remaining to unstack then ultimately , MultiIndex DataFrame converts to MultiIndex Series
#### when no features are remaining to stack then ultimately , MultiIndex DataFrame converts to MultiIndex Series

```python
df.unstack()
```

```python
df.unstack().unstack()
```

```python
df.unstack().unstack().unstack()
```

```python
df.unstack().unstack().stack()
```

```python
df.unstack().unstack().stack().stack()
```

```python
df.unstack().unstack().stack().stack().stack()
```

```python
df.unstack().unstack().stack().stack().stack()
```

```python
df.unstack().unstack().stack().stack().stack().stack()
```

##### we can sort MultiIndex DataFrame on basis of multiple indices using sort_index method as below by specific level of row features

```python
df.sort_index(level=[1])
```

### all methods of DataFrame work similarly for these MultiIndex DataFrames as ultimately they are DataFrames
